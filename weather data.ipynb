Looking at the values, the time columns are the same, one is a timestamp and the other is utc converted.

The coordinates do not contirbute to the data, so we can remove the name, longitude and latitude

rain_3h , snow_3h, sea_level, ground_level, are all nana so we can remove those features.


```python
# dropping columns
drop_titles = ["city_name", "lat", "lon", "sea_level", "grnd_level", "rain_3h", "snow_3h"]
weather_df = weather_df.drop(drop_titles, axis=1)
```

```python
weather_df.dtypes
```

```python
category = ["timezone", "weather_id", "weather_main", "weather_description", "weather_icon", "clouds_all"]
weather_df[category] = weather_df[category].astype("category")
```
